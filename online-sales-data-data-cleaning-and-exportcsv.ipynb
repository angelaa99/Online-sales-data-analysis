```python
import pandas as pd
```

```python
data=pd.read_csv('/kaggle/input/online-sales-dataset/online_sales_dataset.csv')
data.head()
```

```python
# Understand the dataset
print(data.info())
print(data.shape)
print(data.describe())
```

```python
# Identifying any null value
data.isnull().sum()
```

```python
# Drop null values and check the number of row after dropping null values
updated_data = data.dropna(axis=0)
updated_data.info()
```

```python
# Checking any null values still exists
updated_data.isnull().sum()
```

```python
# Deleting unused columns
updated_data = updated_data.drop(['StockCode','CustomerID'],axis=1)
print(updated_data.shape)
```

```python
# checking if the unused columns have been deleted
updated_data['Description'].unique()
```

```python
# Checking any duplicates
updated_data.duplicated().sum()
```

```python
# Export the updated dataset to csv for further analysis using SQL
updated_data.to_csv('online-sales-data.csv', index=False)
```
